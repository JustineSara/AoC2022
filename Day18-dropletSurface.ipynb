```python
X = []
Y = []
Z = []

for l in my_input.split("\n") :
    if l == "":
        continue
    x,y,z = l.split(",")
    X.append(int(x))
    Y.append(int(y))
    Z.append(int(z))
```

```python
M = np.zeros((max(X)+1,max(Y)+1,max(Z)+1))
for x,y,z in zip(X,Y,Z) :
    M[x,y,z] = 1
```

```python
Nadj = 0
for x,y,z in zip(X,Y,Z) :
    if x != M.shape[0]-1 :
        Nadj += M[x+1,y,z]
    if x != 0 :
        Nadj += M[x-1,y,z]
    
    if y != M.shape[1]-1 :
        Nadj += M[x,y+1,z]
    if y != 0 :
        Nadj += M[x,y-1,z]
        
    
    if z != M.shape[2]-1 :
        Nadj += M[x,y,z+1]
    if z != 0 :
        Nadj += M[x,y,z-1]
        
print(Nadj)
print(6*len(X)-(Nadj))
```

```python
M = np.zeros((max(X)+2,max(Y)+2,max(Z)+2))
for x,y,z in zip(X,Y,Z) :
    M[x,y,z] = 1
```

```python
Nadj = 0

steam = [(0,0,0)]

while len(steam) > 0 :
    
    x,y,z = steam.pop(0)
     
    print(x,y,z,M[x,y,z],"\t",len(steam),"\t",Nadj)
    if M[x,y,z] == 10 :
        continue

    if x != M.shape[0]-1 :
        if M[x+1,y,z] == 1 :
            Nadj += 1
        else :
            steam.append((x+1,y,z))
    if x != 0 :
        if M[x-1,y,z] == 1 :
            Nadj += 1
        else : 
            steam.append((x-1,y,z))

    if y != M.shape[1]-1 :
        if M[x,y+1,z] == 1 :
            Nadj += 1
        else :
            steam.append((x,y+1,z))
    if y != 0 :
        if M[x,y-1,z] == 1 :
            Nadj += 1
        else : 
            steam.append((x,y-1,z))

    if z != M.shape[2]-1 :
        if M[x,y,z+1] == 1 :
            Nadj += 1
        else :
            steam.append((x,y,z+1))
    if z != 0 :
        if M[x,y,z-1] == 1 :
            Nadj += 1
        else : 
            steam.append((x,y,z-1))

    M[x,y,z] = 10

print(Nadj)
```
