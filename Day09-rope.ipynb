```python
class Rope :
    
    def __init__(self) :
        self.head_pos = [0,0]
        self.tail_pos = [0,0]
        
    def move_head(self,d) :
        if d == 'R':
            self.head_pos[1] += 1
        elif d == 'L':
            self.head_pos[1] -= 1
        elif d == 'U' :
            self.head_pos[0] += 1
        elif d == 'D' :
            self.head_pos[0] -= 1
        else :
            print("error in direction to move head")

    def move_tail(self):
        if abs(self.tail_pos[1] - self.head_pos[1]) > 1 or abs(self.tail_pos[0] - self.head_pos[0])>1 :
            self.move_one_dim(0)
            self.move_one_dim(1)
            
    def move_one_dim(self,di):
        if self.tail_pos[di] < self.head_pos[di] :
            self.tail_pos[di] += 1
        elif self.tail_pos[di] > self.head_pos[di] :
            self.tail_pos[di] -= 1
    
    def update_head_pos(self,pos):
        self.head_pos = pos
        
    def read_tail_pos(self):
        return (self.tail_pos[0],self.tail_pos[1])
        
    def __str__(self):
        this_str = ''
        for i0 in range( max(5,self.head_pos[0]+1,self.tail_pos[0]+1)+1,  min(-1,self.head_pos[0]-1,self.tail_pos[0]-1) , -1) :
            for i1 in range( min(-1,self.head_pos[1]-1,self.tail_pos[1]-1) , max(5,self.head_pos[1]+1,self.tail_pos[1]+1)+1 ) :
                if [i0,i1] == self.head_pos:
                    this_str += 'H '
                elif [i0,i1] == self.tail_pos:
                    this_str += 'T '
                else :
                    this_str += '. '
            this_str += '\n'
        return this_str
```

```python
def part1(this_input):
    
    r=Rope()
    all_tail_position = set()
    all_tail_position.add((0,0))
    for l in this_input.split("\n"):
        if l=='':
            continue
        [d , n] = l.split(" ")
        n= int(n)
#         print(d,n)
        for i in range(n):
            r.move_head(d)
            r.move_tail()
            all_tail_position.add(r.read_tail_pos())
#         print(r)
        
    return len(all_tail_position) , all_tail_position
```

```python
class fullRope() :

    def __init__(self, N) :
        knots = []
        for i in range(N) :
            knots.append(Rope())
        self.knots = knots
        self.N = N
    
    def move(self,d) :
        self.knots[0].move_head(d)
        self.knots[0].move_tail()
        for i in range(1,self.N):
            self.knots[i].update_head_pos(self.knots[i-1].tail_pos)
            self.knots[i].move_tail()
    
    def get_tail_end(self) :
        return (self.knots[self.N-1].tail_pos[0],self.knots[self.N-1].tail_pos[1])
    
    def __str__(self):
        this_str = ''
        for i0 in range( 15 ,  -6 , -1) :
            for i1 in range( -11 , 15 ) :
                this_l = ". "
                if self.knots[self.N-1].tail_pos == [i0,i1] :
                    this_l = str(self.N)+" "
                for k in range(self.N-1,0,-1) :
                    if [i0,i1] == self.knots[k].head_pos:
                        this_l = str(k)+' '
                if [i0,i1] == self.knots[0].head_pos:
                    this_l = 'H '
                this_str += this_l
            this_str += '\n'
        return this_str
```
