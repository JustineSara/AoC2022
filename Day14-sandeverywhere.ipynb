```python
class cave :
    
    def __init__(self, this_input) :
        
        self.minx, self.miny, self.c = self.create_empty_cave(this_input)
        self.rocks_in_cave(this_input)
        self.sand_source()
        
    def __str__(self) :
        s = ""
        for l in self.c.transpose() :
            for el in l :
                if el == 500 :
                    s += "+"
                elif el == 10 :
                    s += "#"
                elif el == 1 :
                    s += "o"
                else :
                    s += "."
            s+="\n"
        return s

    def create_empty_cave(self, this_input):
        listx = [500]
        listy = [0]
        for l in this_input.split("\n") :
            if l == "" :
                continue
            for p in l.split(" -> ") :
                x,y = p.split(",")
                listx.append(int(x))
                listy.append(int(y))
        minx = min(listx)
        maxx = max(listx)
        miny = min(listy)
        maxy = max(listy)
        lenx = maxx-minx +1
        leny = maxy-miny +1

        c = np.zeros((lenx,leny))

        return minx, miny, c
    
    def text_to_pos(self,p):
        x,y = p.split(",")
        return (int(x)-self.minx , int(y)-self.miny)
    
    def rocks_in_cave(self,this_input):
        for l in this_input.split("\n") :
            if l == "" :
                continue
            p = l.split(" -> ")
            if len(p) == 1 :
                self.c[self.text_to_pos(p)] = 10
            for i in range(len(p)-1) :
                x1,y1 = self.text_to_pos(p[i])
                x2,y2 = self.text_to_pos(p[i+1])
                if x1 == x2 :
                    for y in range(min(y1,y2),max(y1,y2)+1) :
                        self.c[x1,y] = 10
                elif y1==y2 :
                    for x in range(min(x1,x2),max(x1,x2)+1) :
                        self.c[x,y1] = 10
                else :
                    print("problem drawing line of rock : p1 =",x1,y1,"  - p2 =",x2,y2)
        return True
    
    def sand_source(self) :
        self.c[500-self.minx, 0-self.miny] = 500
    
    
    def sand_flow(self) :
        sx = 500-self.minx
        sy = 0-self.miny
        can_flow = True
        
        while(can_flow) :
            
            if sy + 1 >= self.c.shape[1] :
                return False
            elif self.c[sx,sy+1] == 0 :
                sy += 1
            elif sx-1 < 0 :
                return False
            elif self.c[sx-1,sy+1] == 0 :
                sx -= 1
                sy += 1
            elif sx+1 >= self.c.shape[0] :
                return False
            elif self.c[sx+1,sy+1] == 0 :
                sx += 1
                sy += 1
            else :
                can_flow = False
        
        self.c[sx,sy] = 1
        return True
    
    def part1(self):
        Nsands = 0
        while(self.sand_flow()) :
            Nsands += 1
        return Nsands
```

```python
class cave2 :
    
    def __init__(self, this_input) :
        
        self.minx, self.miny, self.c = self.create_empty_cave(this_input)
        self.rocks_in_cave(this_input)
        self.sand_source()
        
    def __str__(self) :
        s = ""
        for l in self.c.transpose() :
            for el in l :
                if el == 500 :
                    s += "+"
                elif el == 10 :
                    s += "#"
                elif el == 1 :
                    s += "o"
                else :
                    s += "."
            s+="\n"
        return s

    def create_empty_cave(self, this_input):
        listx = [500]
        listy = [0]
        for l in this_input.split("\n") :
            if l == "" :
                continue
            for p in l.split(" -> ") :
                x,y = p.split(",")
                listx.append(int(x))
                listy.append(int(y))
        minx = min(listx)
        maxx = max(listx)
        miny = min(listy)
        maxy = max(listy)
        lenx = maxx-minx +1
        leny = maxy-miny +1 +2

        c = np.zeros((lenx,leny))

        return minx, miny, c
    
    def text_to_pos(self,p):
        x,y = p.split(",")
        return (int(x)-self.minx , int(y)-self.miny)
    
    def rocks_in_cave(self,this_input):
        for l in this_input.split("\n") :
            if l == "" :
                continue
            p = l.split(" -> ")
            if len(p) == 1 :
                self.c[self.text_to_pos(p)] = 10
            for i in range(len(p)-1) :
                x1,y1 = self.text_to_pos(p[i])
                x2,y2 = self.text_to_pos(p[i+1])
                if x1 == x2 :
                    for y in range(min(y1,y2),max(y1,y2)+1) :
                        self.c[x1,y] = 10
                elif y1==y2 :
                    for x in range(min(x1,x2),max(x1,x2)+1) :
                        self.c[x,y1] = 10
                else :
                    print("problem drawing line of rock : p1 =",x1,y1,"  - p2 =",x2,y2)
        
        y = self.c.shape[1]-1
        for x in range(self.c.shape[0]) :
            self.c[x,y] = 10
        return True
    
    def sand_source(self) :
        self.c[500-self.minx, 0-self.miny] = 500
    
    
    def sand_flow(self) :
        sx = 500-self.minx
        sy = 0-self.miny
        can_flow = True
        
        while(can_flow) :
            
            if sy + 1 >= self.c.shape[1] :
                print("this should not happen anymore!")
                return False
            elif self.c[sx,sy+1] == 0 :
                sy += 1
            elif sx-1 < 0 :
                # add a column
                N = self.c.shape[1] 
                self.c = np.r_[ np.zeros((1,N)) , self.c ]
                self.minx -= 1
                self.c[0,N-1] = 10
                self.c[0,N-2] = 1
                return True
            elif self.c[sx-1,sy+1] == 0 :
                sx -= 1
                sy += 1
            elif sx+1 >= self.c.shape[0] :
                # add a column
                N = self.c.shape[1] 
                self.c = np.r_[ self.c , np.zeros((1,N))  ]
                self.c[self.c.shape[0]-1,N-1] = 10
                self.c[self.c.shape[0]-1,N-2] = 1
                return True
            elif self.c[sx+1,sy+1] == 0 :
                sx += 1
                sy += 1
            else :
                can_flow = False
        if sx == 500-self.minx and sy == 0-self.miny :
            return False
        self.c[sx,sy] = 1
        return True
    
    def part2(self):
        Nsands = 0
        while(self.sand_flow()) :
            Nsands += 1
        return Nsands + 1
    
```
